````python
import pandas as pd
import numpy as np

np.random.seed(0)

products = ['apple', 'banana', 'cherry', 'dates', 'elderberry']
categories = ['fruit', 'fruit', 'fruit', 'fruit', 'fruit']

N_ROWS = 10**6
df = pd.DataFrame({
    'product_name': np.random.choice(products, N_ROWS),
    'product_category': np.random.choice(categories, N_ROWS),
    'product_price': np.random.uniform(1, 10, N_ROWS),
    'transaction_date': pd.date_range(start='1/1/2000', periods=N_ROWS, freq='h')
})

# Пам’ять до перетворення
memory_before = df.memory_usage(deep=True).sum()
print(f"Memory usage before: {memory_before / 1024**2:.2f} MB")

# Перетворення на категорійний тип
df['product_name'] = df['product_name'].astype('category')
df['product_category'] = df['product_category'].astype('category')

# Пам’ять після перетворення
memory_after = df.memory_usage(deep=True).sum()
print(f"Memory usage after: {memory_after / 1024**2:.2f} MB")

# Економія пам’яті
reduction_factor = memory_before / memory_after
print(f"Memory reduced by factor: {reduction_factor:.2f}x")
````

````python
def calculate_variance(numbers):
    """Обчислює дисперсію списку чисел."""
    total = sum(numbers)
    mean = total / len(numbers)
    
    total_squares = sum(n**2 for n in numbers)
    mean_squares = total_squares / len(numbers)
    
    variance = mean_squares - mean**2
    return variance


input_list = [75, 80, 85, 90, 95]
result = calculate_variance(input_list)
print(f"Variance: {result}")
````
