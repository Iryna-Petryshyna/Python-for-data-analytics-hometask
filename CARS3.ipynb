```python
# Імпортуємо бібліотеки
import pandas as pd
import numpy as np
```

```python
# Завантажуємо і об'єднуємо файли в один
files = [
    "tz_opendata_z01012022_po01012023.csv",
    "tz_opendata_z01012023_po01012024.csv",
    "tz_opendata_z01012024_po01012025.csv"
]

dfs = [pd.read_csv(f, sep=";", encoding="utf-8") for f in files]
df_all = pd.concat(dfs, ignore_index=True)

print(df_all.shape)

# Виводимо перших 5 рядків
df_all.head()
```

Повідомлення DtypeWarning. 
У різних файлах один і той самий стовпець має різні типи (наприклад, у одному число, у другому текст). У стовпцях 17 і 18 в різних частинах файлу pandas зустрів різні типи (наприклад, числа та текст).
У такому випадку pandas читає все як object (рядки). Тому далі виводимо усі типи даних і приводимо до правильного формату.

```python
# Перевіряємо типи даних у злитому даатсеті
df_all.info()
```

**Висовок по результатах аналізу типів даних:**

Аналізуючи типи даних бачимо, ряд проблем:
- OWN_WEIGHT і TOTAL_WEIGHT мають тип даних object, а мають мати тип даних числовий. 
- OPER_NAME, DEP, BRAND, MODEL, COLOR, KIND, BODY, PURPOSE, FUEL - краще перевести до category. Бо вони не будуть обчислюватись, а тільки групуватись/фільтруватись. Це зменить об'єм пам'яті. 
- D_REG має тип даних object, краще перевести до datetime. 
Тому далі переводиво ці дві колонки до числового типу.

```python
# Перетворення OWN_WEIGHT і TOTAL_WEIGHT у числовий формат
df_all["OWN_WEIGHT"] = pd.to_numeric(df_all["OWN_WEIGHT"], errors="coerce")
df_all["TOTAL_WEIGHT"] = pd.to_numeric(df_all["TOTAL_WEIGHT"], errors="coerce")
```

```python
# Переводимо категоріальні дані до category
cat_columns = [
    "OPER_NAME", "DEP", "BRAND", "MODEL", "COLOR",
    "KIND", "BODY", "PURPOSE", "FUEL"
]
for col in cat_columns:
    df_all[col] = df_all[col].astype("category")
```

```python
# Переводимо D_REG в datetime, щоб Tableau одразу коректно його розпізнав.
df_all["D_REG"] = pd.to_datetime(df_all["D_REG"], errors="coerce")
```

```python
# Вказуємо одразу формат дати
df_all["D_REG"] = pd.to_datetime(df_all["D_REG"], format="%d.%m.%Y", errors="coerce")
```

```python
# Кількість некоректно розпізнаних дат (NaT)
invalid_dates = df_all["D_REG"].isna().sum()
print(f"Кількість некоректних дат: {invalid_dates}")

# Переглянути приклади некоректних рядків
print(df_all[df_all["D_REG"].isna()][["D_REG"]].head(10))

# Мінімальна та максимальна дата
print("Мінімальна дата:", df_all["D_REG"].min())
print("Максимальна дата:", df_all["D_REG"].max())

# Скільки різних років є в даних
print("Унікальні роки:", df_all["D_REG"].dt.year.unique())

# Розподіл по роках (для наглядності)
print(df_all["D_REG"].dt.year.value_counts().sort_index())
```

```python
# Перевіряємо оновлені типи даних у злитому даатсеті
df_all.info()
```

```python
# Додаємо технічну колонку рік набору даних. Це допоможе в Tableau будувати зріз по роках.

files = {
    "tz_opendata_z01012022_po01012023.csv": 2022,
    "tz_opendata_z01012023_po01012024.csv": 2023,
    "tz_opendata_z01012024_po01012025.csv": 2024
}

dfs = []

for f, year in files.items():
    df_temp = pd.read_csv(
        f,
        sep=";",
        encoding="utf-8",
        dtype={
            "OWN_WEIGHT": "string",
            "TOTAL_WEIGHT": "string"
        },
        low_memory=False
    )
    df_temp["YEAR"] = year
    dfs.append(df_temp)

df_all = pd.concat(dfs, ignore_index=True)

# тепер приводимо вагу до числового типу
df_all["OWN_WEIGHT"] = pd.to_numeric(df_all["OWN_WEIGHT"], errors="coerce")
df_all["TOTAL_WEIGHT"] = pd.to_numeric(df_all["TOTAL_WEIGHT"], errors="coerce")

print(df_all.shape)
print(df_all.dtypes)
```

```python
# Створюємо колонку вік авто на момент року реєстрації
df_all['CAR_AGE'] = df_all['YEAR'] - df_all['MAKE_YEAR']

# Якщо є негативні значення або пропуски — замінюємо на NaN
df_all['CAR_AGE'] = df_all['CAR_AGE'].apply(lambda x: x if x >= 0 else pd.NA)

# Перевірка
df_all[['MAKE_YEAR', 'YEAR', 'CAR_AGE']].head()
```

```python
# Створюємо категорії віку авто за роками
bins = [-1, 1, 5, 10, 20, 100]  # верхня межа 100 років
labels = ['0-1', '2-5', '6-10', '11-20', '21+']

df_all['AGE_CATEGORY'] = pd.cut(df_all['CAR_AGE'], bins=bins, labels=labels)

# Перевірка
df_all[['MAKE_YEAR', 'YEAR', 'CAR_AGE', 'AGE_CATEGORY']].head(10)
```

```python
# Аналіз пропущених значень
missing_data = df_all.isnull().sum()                        # Кількість пропусків
missing_percent = (missing_data / len(df_all)) * 100       # Відсоток пропусків

# Створюємо зручну таблицю
missing_summary = pd.DataFrame({
    'Пропущені': missing_data,
    '%': missing_percent.round(2)  # округлення до сотих
}).sort_values(by='%', ascending=False)  # сортуємо від найбільш проблемних колонок

print("Пропущені значення (кількість і %):")
print(missing_summary)
```

```python
# Перетворюємо КОАТУУ в рядок, щоб уникнути наукової нотації
df_all["REG_ADDR_KOATUU"] = df_all["REG_ADDR_KOATUU"].astype("Int64").astype(str)

# Довідник: перші дві цифри → область
koatuu_to_region = {
    "01": "АР Крим",
    "05": "Вінницька",
    "07": "Волинська",
    "12": "Дніпропетровська",
    "14": "Донецька",
    "18": "Житомирська",
    "21": "Закарпатська",
    "23": "Запорізька",
    "26": "Івано-Франківська",
    "32": "Київська",
    "35": "Кіровоградська",
    "44": "Луганська",
    "46": "Львівська",
    "48": "Миколаївська",
    "51": "Одеська",
    "53": "Полтавська",
    "56": "Рівненська",
    "59": "Сумська",
    "61": "Тернопільська",
    "63": "Харківська",
    "65": "Херсонська",
    "68": "Хмельницька",
    "71": "Черкаська",
    "73": "Чернівецька",
    "74": "Чернігівська",
    "80": "м. Київ",
    "85": "м. Севастополь",
}

# Витягаємо перші дві цифри з коду
df_all["REGION_CODE"] = df_all["REG_ADDR_KOATUU"].str[:2]

# Додаємо назву області
df_all["REGION"] = df_all["REGION_CODE"].map(koatuu_to_region)
```

```python
# Виводимо унікальні значення REGION
unique_ops = df_all['REGION'].unique()

print("Унікальні операції (REGION):")
for op in unique_ops:
    print(op)
```

**Висновки із аналізу пропущених значень:**
- CAPACITY (7.73%) – важлива характеристика для аналізу трендів, тому пропуски заповнюємо NaN.
- FUEL (4.61%) – важлива характеристика для аналізу трендів, тому пропуски заповнюємо NaN.
- N_REG_NEW (1.08%) – мало, залишаємо як є, оскілька державний номерний знак ТЗ не є суттєвими для аналізу трендів.
- в інших відсоток пропущених значень малий, можемо ігнорувати.

```python
# Замінюємо порожні рядки або пробіли на NaN у колонці FUEL і CAPACITY
df_all['FUEL'] = df_all['FUEL'].replace(r'^\s*$', np.nan, regex=True)
df_all['CAPACITY'] = df_all['CAPACITY'].replace(r'^\s*$', np.nan, regex=True)

# Перевірка
print(df_all[['FUEL','CAPACITY']].isna().sum())
```

```python
# Загальна кількість реєстрацій
total_regs = len(df_all)

# Середня кількість на місяць
regs_per_month = (
    df_all.groupby(df_all['D_REG'].dt.to_period('M'))['D_REG']
    .count()
    .mean()
)

# Найпопулярніший бренд
top_brand = df_all['BRAND'].value_counts().idxmax()

# Найпопулярніший колір
top_color = df_all['COLOR'].value_counts().idxmax()

# Частка електроавто
share_ev = (
    df_all['FUEL'].str.contains('електр', case=False, na=False).sum()
    / len(df_all)
) * 100

# Найпоширеніше пальне
top_fuel = df_all['FUEL'].value_counts().idxmax()

# === Вивід результатів ===
print("Основні KPI")
print(f"Загальна кількість реєстрацій: {total_regs:,}")
print(f"Середня кількість на місяць: {regs_per_month:.0f}")
print(f"Найпопулярніший бренд: {top_brand}")
print(f"Найпопулярніший колір: {top_color}")
print(f"Частка електроавто: {share_ev:.2f}%")
print(f"Найпоширеніше пальне: {top_fuel}")
```

```python
# Загальна кількість реєстрацій
total_registrations = len(df_all)
print(f"Загальна кількість реєстрацій: {total_registrations}")

# Середня кількість на місяць
# Спершу приводимо колонку D_REG до datetime
df_all['D_REG'] = pd.to_datetime(df_all['D_REG'], errors='coerce')

# Визначаємо період
period_months = ((df_all['D_REG'].max() - df_all['D_REG'].min()).days / 30.44)
avg_per_month = total_registrations / period_months
print(f"Середня кількість реєстрацій на місяць: {avg_per_month:.2f}")

# Найпопулярніший бренд
top_brand = df_all['BRAND'].value_counts().idxmax()
top_brand_count = df_all['BRAND'].value_counts().max()
top_brand_pct = top_brand_count / total_registrations * 100
print(f"Найпопулярніший бренд: {top_brand} ({top_brand_count} реєстрацій, {top_brand_pct:.2f}%)")

# Найпопулярніший колір
top_color = df_all['COLOR'].value_counts().idxmax()
top_color_count = df_all['COLOR'].value_counts().max()
top_color_pct = top_color_count / total_registrations * 100
print(f"Найпопулярніший колір: {top_color} ({top_color_count} реєстрацій, {top_color_pct:.2f}%)")

# Частка електроавто (припускаємо, що KIND або FUEL містить "Electric" чи подібне)
if 'FUEL' in df_all.columns:
    electric_count = df_all[df_all['FUEL'].str.contains('electro|electric', case=False, na=False)].shape[0]
    electric_pct = electric_count / total_registrations * 100
    print(f"Частка електроавто: {electric_count} ({electric_pct:.2f}%)")

# Найпоширеніше пальне
top_fuel = df_all['FUEL'].value_counts().idxmax()
top_fuel_count = df_all['FUEL'].value_counts().max()
top_fuel_pct = top_fuel_count / total_registrations * 100
print(f"Найпоширеніше пальне: {top_fuel} ({top_fuel_count} реєстрацій, {top_fuel_pct:.2f}%)")
```
